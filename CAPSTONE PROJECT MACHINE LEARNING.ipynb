```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
fandango = pd.read_csv("samplemain.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(data=fandango, x='RATING')
plt.show()
```

```python
tp = fandango.corr()
print(tp)
sp = fandango.nlargest(10,'VOTES')
print(sp)
```

```python
sns.kdeplot(data=fandango, x='RATING',clip=[0,5],fill=True,label='True Rating')
sns.kdeplot(data=fandango, x='STARS',clip=[0,5],fill=True,label='STARS Display')
plt.legend(loc=(1.05,0.5))
plt.figure(figsize=(10,4),dpi=150)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("samplefilm.csv")
print(df.head(5))
plt.figure(dpi=150)
sns.scatterplot(data=df,x='RottenTomatoes',y='RottenTomatoes_User')
plt.show()
```

```python
sns.countplot(data=df,x='RottenTomatoes',palette = 'magma')
plt.show()
```

```python
df['Rotten_Diff'] = df['RottenTomatoes'] -df['RottenTomatoes_User']
```

```python
df['Rotten_Diff'].apply(abs).mean()
```

```python
plt.figure(dpi=150)
sns.histplot(data=df,x='Rotten_Diff',kde=True,bins=25)
plt.show()
```

```python
df.nsmallest(5,'Rotten_Diff')
```

```python
plt.figure(dpi=150)
sns.scatterplot(data=df,x='Metacritic',y='Metacritic_User')
plt.show()
```

```python
sns.countplot(data=df,x='Metacritic')
plt.show()
```

```python
df['Metacritic_Diff'] = df['Metacritic'] -df['Metacritic_User']
df['Metacritic_Diff'].apply(abs).mean()
```

```python
plt.figure(dpi=150)
sns.histplot(data=df,x='Metacritic_Diff',kde=True,bins=25)
plt.show()
```
